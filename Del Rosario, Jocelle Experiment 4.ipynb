# EXPERIMENT 4 - Data Wrangling and Data Visualization.

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
#read the file
df = pd.read_csv('board2.csv')
df
```

```python
# make the visayas as constant and it should not be included in the table
Vis = df[(df['Hometown'] == 'Visayas') & (df['Math']<70)]
```

```python
# list the needed data
Vis = Vis[['Name', 'Gender', 'Track', 'Math']]
Vis
```

```python
#plot the figure
plt.figure(figsize=(6,6))
plt.bar(df['Track'], df['Math'], data='Vis')
plt.title('The Students Track in Visayas with Math Score<70')
plt.show()
```

```python
#make the luzon and intrumentation as constant and it should not be included in the table
Instru = df[(df['Track'] == 'Instrumentation') & (df['Hometown'] == 'Luzon')& (df['Electronics']>70)]
```

```python
# list the needed data
Instru = Instru[['Name', 'GEAS', 'Electronics']]
Instru
```

```python
import numpy as np
# plot the given table, it should include the geas and electronics
index = np.arange(len(Instru))
bar_width = 0.2
plt.figure(figsize=(7,7))
plt.bar(index, (Instru['GEAS']), bar_width, label='GEAS', color='blue')
plt.bar(index + bar_width,Instru['Electronics'], bar_width, label='Electronics', color='red')

plt.title('The Score in GEAS of Instrumentation Students from Luzon with Electronics>70')
plt.xlabel('student name')
plt.ylabel('student scores')
plt.xticks([i + bar_width / 2 for i in index],Instru['Name'])
plt.legend(loc='upper right')
```

```python
# make a data that has an average of the subjects
df['Average'] = df[['Math', 'GEAS', 'Electronics', 'Communication']].mean(axis=1)
df
```

```python
# female and mindanao are constant and should not be included on the table
Mindy = df[(df['Gender'] == 'Female') & (df['Hometown'] == 'Mindanao')& (df['Average']>=55)]
```

```python
# list the needed data
Mindy = Mindy[['Name', 'Track', 'Electronics','Average']]
Mindy
```

```python
#plot the table
index = np.arange(len(Mindy))
bar_width = 0.2
plt.figure(figsize=(7,7))
plt.bar(index, (Mindy['Average']), bar_width, label='Average', color='brown')
plt.bar(index + bar_width,Mindy['Electronics'], bar_width, label='Electronics', color='green')
#label the x and y
plt.title('The Score in Electronics of Female Students in Mindanao who has an average of >=55')
plt.xlabel('student name')
plt.ylabel('student scores')
plt.xticks([i + bar_width / 2 for i in index], Mindy['Name'])
plt.legend(loc='upper right')
```

```python
# get the average
df['Average'] = df[['Math', 'GEAS', 'Electronics', 'Communication']].mean(axis=1)

# Group by track, gender, and hometown and calculate the mean of the average scores
trackavg = df.groupby('Track')['Average'].mean().reset_index()
genderavg = df.groupby('Gender')['Average'].mean().reset_index()
hometownavg = df.groupby('Hometown')['Average'].mean().reset_index()
```

```python
import seaborn as sns
plt.figure(figsize=(15, 5))

# Plot 1: Track vs Average
plt.subplot(1, 3, 1)
sns.barplot(x='Track', y='Average', data=trackavg, color='Blue')
plt.title('Average Grade by Track')
plt.xlabel('Track')
plt.ylabel('Average Score')

# Plot 2: Gender vs Average
plt.subplot(1, 3, 2)
sns.barplot(x='Gender', y='Average', data=genderavg, color='Green')
plt.title('Average Grade by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Score')

# Plot 3: Hometown vs Average
plt.subplot(1, 3, 3)
sns.barplot(x='Hometown', y='Average', data=hometownavg, color='Red')
plt.title('Average Grade by Hometown')
plt.xlabel('Hometown')
plt.ylabel('Average Score')

#show
plt.tight_layout()
plt.show()
```
